## problem2

$(a)$ It is easily obeserved that two curves are tangent to each other when alpha = 0, codes and figures shown below

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as random
import numpy.linalg as la
import warnings
warnings.filterwarnings("ignore")
```

```python
#x here is a vector of 5 dims
def f(x):
    return np.log(np.sum(np.exp(x)))

#x here is a vector of 5 dims
def gradF(x):
    return np.exp(x) / np.sum(np.exp(x))

random.seed(0)
x = random.randn(5)
v0 = random.randn(5)
v1 = random.randn(5)

def gradPlot(v=v0):
    alpha = np.linspace(-1, 1, 100)  
    if isinstance(v, list):
        fig, ax = plt.subplots(figsize=(8, 6))
        for v_ in v[1:]:
            fInShift = [f(x + alp*v_) for alp in alpha]
            ax.plot(alpha, fInShift, linestyle='--', label=r"$v\ != \nabla f(x)$")
        ax.plot(alpha, [f(x + alp*v[0]) for alp in alpha], label=r"$v=\nabla f(x)$", lw=3)   
        ax.plot(alpha, [f(x) + np.dot(alp*gradF(x).T,v[0]) for alp in alpha], label="tangent", lw=2)
        ax.set_xlabel(r"$\alpha$")
        ax.legend()
           
    else:
        fInShift = [f(x + alp*v) for alp in alpha]
        fOtShift = [f(x) + np.dot(alp*gradF(x).T,v) for alp in alpha]
        fig, ax = plt.subplots(figsize=(5, 4))
        ax.plot(alpha, fInShift, label=r"$f(x + \alpha v)$")
        ax.plot(alpha, fOtShift, label=r"$f(x)+\alpha(\nabla f(x))^Tv$")
        ax.set_xlabel(r"$\alpha$")
        ax.legend()
```

```python
gradPlot(v0)
gradPlot(v1)
```

$(b)$ When v = gradient of f(x), two curves are very close and the curve with v = gradient increases fastest.

```python
random.seed(3)
v0 = gradF(x)
v = [v0, random.randn(5), random.randn(5)]
gradPlot(v=v)
```
